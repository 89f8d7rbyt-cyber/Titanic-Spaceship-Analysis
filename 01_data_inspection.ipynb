# Spaceship Titanic - Data Inspection

This notebook loads and inspects `train.csv`, `test.csv`, and `sample_submission.csv`.

```python
# Import Path to manage file locations robustly across environments.
from pathlib import Path
# Import pandas for CSV loading and DataFrame inspection.
import pandas as pd
```

```python
# Resolve base directory: if running inside notebooks/, move to project root; otherwise use current directory.
base_dir = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
# Build path for train.csv.
train_path = base_dir / 'train.csv'
# Build path for test.csv.
test_path = base_dir / 'test.csv'
# Build path for sample_submission.csv.
sample_submission_path = base_dir / 'sample_submission.csv'

# Load train dataset into memory.
train = pd.read_csv(train_path)
# Load test dataset into memory.
test = pd.read_csv(test_path)
# Load sample submission template into memory.
sample_submission = pd.read_csv(sample_submission_path)

# Print resolved directory to confirm where files were loaded from.
print('Loaded files from:', base_dir)
```

```python
# Define a reusable function to print key quality checks for any DataFrame.
def inspect_df(name, df, include_target=False):
    # Print separator line for section readability.
    print('=' * 80)
    # Print dataset label.
    print(f'Dataset: {name}')
    # Print closing separator for header block.
    print('=' * 80)

    # Print row/column count as first overview metric.
    print('Shape:', df.shape)

    # Print label for dtype section.
    print('Dtypes:')
    # Print pandas-inferred data types for each column.
    print(df.dtypes)

    # Print label for missing-value section.
    print('Missing values per column:')
    # Print missing value counts sorted descending.
    print(df.isna().sum().sort_values(ascending=False))

    # Print label for duplicate checks.
    print('Duplicates:')
    # Print number of fully duplicated rows.
    print('- Duplicate rows:', df.duplicated().sum())
    # If PassengerId exists, print duplicate PassengerId count.
    if 'PassengerId' in df.columns:
        print('- Duplicate PassengerId:', df['PassengerId'].duplicated().sum())

    # Optionally print target distribution for training data analysis.
    if include_target and 'Transported' in df.columns:
        # Print class-balance section title.
        print('Class balance (Transported):')
        # Count each class value including potential missing labels.
        counts = df['Transported'].value_counts(dropna=False)
        # Compute normalized class proportions.
        ratios = df['Transported'].value_counts(normalize=True, dropna=False)
        # Print counts and ratios side by side.
        print(pd.concat([counts.rename('count'), ratios.rename('ratio')], axis=1))

    # Print blank line to separate dataset outputs.
    print()
```

```python
# Run inspection on train dataset and include target class balance.
inspect_df('train.csv', train, include_target=True)
# Run inspection on test dataset.
inspect_df('test.csv', test)
# Run inspection on sample submission dataset.
inspect_df('sample_submission.csv', sample_submission)
```
